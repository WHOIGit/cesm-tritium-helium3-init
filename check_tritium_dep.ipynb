```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import netCDF4
from matplotlib import colors
%matplotlib notebook
```

```python
fp = netCDF4.Dataset('/home/ivan/Python/data/gx1v6.nc','r')
tlat = fp.variables['TLAT'][:]
tlon = fp.variables['TLONG'][:]
fp.close()
nlon, nlat = 320, 384
tritium_dep = np.loadtxt('dep.txt')
tritium_dep = tritium_dep.reshape((nlat,nlon))
tritium_dep.min(), tritium_dep.max()
```

```python
fig, ax = plt.subplots()
im = ax.imshow(tritium_dep,origin='lower',cmap=plt.cm.OrRd,norm=colors.LogNorm())
ax.set(title='[$^3$H] in precipitation')
cb = fig.colorbar(im,ax=ax)
cb.set_label('TU')
```
